```python
def parselist(inputlist):
    start = inputlist[-1]
    reps = [x.split(" => ") for x in inputlist[:-2]]
    return start,reps
```

#### 1.2 Part 1

```python
def part1(inputlist):
    mols = set([])
    start,reps = parselist(inputlist)
    for rep in reps:
        for ii in range(0,len(start)+1-len(rep[0])):
            newmol = start[:ii]+start[ii:].replace(rep[0],rep[1],1)
            if newmol != start:
                mols.add(newmol)
    # print(mols)
    return len(mols)
```
