#### 1.1 helper functions

```python
def parselist(inputlist):
    num_list = []
    for inst in inputlist:
        if inst[0] == 'L':
            num_list.append(-int(inst[1:]))
        else:
            num_list.append(int(inst[1:]))
    return num_list
```

#### 1.2 Part 1

```python
def part1(inputlist):
    dial = 50
    password=0
    for x in parselist(inputlist):
        dial = (dial + x) % 100
        if dial == 0:
            password = password+1
        # print(x,dial,password)
    return password
```
