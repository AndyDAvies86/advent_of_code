```python
def startpoint(inputlist):
    return inputlist[0].index('S')
```

```python
def nextline(instr,lasers):
    newlasers = set([])
    splits = 0
    for laser in lasers:
        if instr[laser] == "^":
            newlasers.add(laser-1)
            newlasers.add(laser+1)
            splits = splits+1
        else:
            newlasers.add(laser)
    return newlasers,splits
```

```python
def nextlinepart2(instr,lasers):
    newlasers = {x:0 for x in range(0,len(instr))}
    splits = 0
    for laser in lasers:
        # print(laser, laser[-1],instr[laser[-1]])
        if instr[laser] == "^":
            newlasers[laser+1] = newlasers[laser+1]+lasers[laser]
            newlasers[laser-1] = newlasers[laser-1]+lasers[laser]
            splits = splits+1
        else:
            newlasers[laser] = newlasers[laser]+lasers[laser]

    return newlasers,splits
```

#### 1.2 Part 1

```python
def part1(inputlist):
    beignat = startpoint(inputlist)
    lasers = set([(beignat)])
    tot_splits = 0
    for line in inputlist:
        lasers,newsplits = nextline(line,lasers)
        tot_splits = tot_splits+newsplits
    return tot_splits
```

```python
def part2(inputlist):
    lasers = {x:0 for x in range(0,len(inputlist[0]))}
    beignat = startpoint(inputlist)
    lasers[beignat] = 1
    tot_splits = 0
    for line in inputlist:
        # print(lasers)
        lasers,newsplits = nextlinepart2(line,lasers)
        tot_splits = tot_splits+newsplits
    return sum(lasers.values())
```
