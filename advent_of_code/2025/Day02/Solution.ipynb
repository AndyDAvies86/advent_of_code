#### 1.1 helper functions

```python
def parselist(inputlist):
    rangelist = [[int(x) for x in y.split("-")] for y in inputlist[0].split(",")]
    # return sorted(rangelist,key=itemgetter(0))   
    return rangelist
```

```python
parselist(test)
```

```python
def checkrange(minmax,n):
    min_val = min(minmax)
    max_val = max(minmax)
    len_min = len(str(min_val))//n
    len_max = -(len(str(max_val))//-n)
    # print(len_min,len_max)
    max_below = (min_val-1)//sum([10**(y*len_min) for y in range(0,n)])
    largest_in = max_val//sum([10**(y*len_max) for y in range(0,n)])
    if max_below > 10**(n*len_min-1):
        max_below = 10**len_min-1
    if largest_in < 10**(len_max-1) and max_val >= 10**(n*(len_min-1))-2 :
        largest_in = 10**(len_max-1) - 1
    # print(f"min:{min_val}, max:{max_val}, len_min:{len_min}, len_max:{len_max}, start:{max_below+1}, end:{largest_in+1}")
    return max_below+1,largest_in+1
```

#### 1.2 Part 1

```python
def part1(inputlist):
    ranges = parselist(inputlist)
    roll_sum = 0
    for chunk in ranges:
        # print(chunk)
        
        # print(checkrange(chunk))
        for ii in range(checkrange(chunk,2)[0],checkrange(chunk,2)[1]):
            inv_id = ii*(10**len(str(ii))+1)
            # print(f"i{inv_id}")
            roll_sum = roll_sum + inv_id
    # print(f"Rolling:{roll_sum}")
    return roll_sum
```

```python
print(f"Part 1 test is: {part1(test)}")
print(f"Part 1 puzzle is: {part1(start)}")
```

#### 1.3 Part 2

```python
def part2(inputlist):
    ranges = parselist(inputlist)
    roll_set = set([])
    for chunk in ranges:
        # print(chunk)
        
        for jj in range (2,15):
            for ii in range(checkrange(chunk,jj)[0],checkrange(chunk,jj)[1]):
                # if ii == checkrange(chunk,jj)[0]:
                    # print(f"n{jj},{checkrange(chunk,jj)}")
                inv_id = ii*sum([10**(y*len(str(ii))) for y in range(0,jj)])
                # print(f"n{jj}, i{inv_id}")
                roll_set.add(inv_id)
    # print(f"Rolling:{roll_sum}")
    return sum(roll_set)
```

```python
print(f"Part 2 test is: {part2(test)}")
print(f"Part 2 puzzle is: {part2(start)}")
```
