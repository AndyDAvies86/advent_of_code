#### 1.1 helper functions

```python
def parselist(inputlist):
    sums_list = []
    rem_space = [x.replace("  "," ").replace("  "," ").replace("  "," ").replace("  "," ").replace("  "," ").replace("  "," ").replace("  "," ").rstrip().lstrip().split(" ") for x in inputlist]
    for ii in range(0,len(rem_space[0])):
        sums_list.append([int(rem_space[x][ii]) for x in range(0,len(rem_space)-1)]+[rem_space[-1][ii]])
    return sums_list
```

```python
def resolve(sum_list):
    # print(sum_list)
    if sum_list[-1] == '+':
        # print(sum(sum_list[:-1]))
        return sum(sum_list[:-1])
    elif sum_list[-1] == '*':
        prod = 1
        for ii in range(0,len(sum_list)-1):
            prod = prod*sum_list[ii]
        # print(prod)
        return prod
    raise Exception
```

```python
def part2parse(inputlist):
    starts=[]
    for ii in range(0,len(inputlist[-1])):
        if inputlist[-1][ii] != " ":
            starts.append(ii)
    starts.append(len(inputlist[-1]))
    # print(starts)
    sums_list = []
    for jj in range(0,len(starts)-1):
        new_sum = []
        for kk in range(starts[jj],starts[jj+1]):
            # print(f"length {len(inputlist[0])}, kk {kk}, jj{jj}, starts {starts[jj]} & {starts[jj+1]}")
            new_num = ''.join([inputlist[x][kk] for x in range(0,len(inputlist)-1)])
            try:
                new_sum.append(int(new_num))
            except:
                pass
        new_sum.append(inputlist[-1][starts[jj]])
        sums_list.append(new_sum)
    return sums_list
```

```python
part2parse(start)
```

#### 1.2 Part 1

```python
def part1(inputlist):
    return sum(resolve(x) for x in parselist(inputlist))
```

#### 1.3 Part 2

```python
def part2(inputlist):
    return sum(resolve(x) for x in part2parse(inputlist))
```
