#### 1.1 helper functions

```python
def parselist(inputlist):
    fresh_rng = [x for x in inputlist[0].split("\n")]
    ingreds = [int(x) for x in inputlist[1].split("\n")]
    return fresh_rng,ingreds
```

```python
def ing_check(ing,fresh_rng):
    for f_r in fresh_rng:
        f_r_int = [int(x) for x in f_r.split("-")]
        if ing >= f_r_int[0] and ing <= f_r_int[1]:
            return 1
    return 0
```

```python
def part1(inputlist):
    fresh_rng,ingreds=parselist(inputlist)
    fresh_count = 0
    for ing in ingreds:
        fresh_count = fresh_count+ing_check(ing,fresh_rng)
    return fresh_count
```

```python
def part2(inputlist):
    fresh_rng,ingreds=parselist(inputlist)
    fresh_rng_int = [[int(x.split("-")[0]),int(x.split("-")[1])] for x in fresh_rng]
    sort_fri = sorted(fresh_rng_int, key=lambda x: x[0]) 
    f_count = sort_fri[0][1]-sort_fri[0][0]+1
    max_fi = sort_fri[0][1]
    for ii in range(1,len(sort_fri)):
        if sort_fri[ii][0] > max_fi:
            f_count = f_count + sort_fri[ii][1]-sort_fri[ii][0]+1
            max_fi = sort_fri[ii][1]
        elif sort_fri[ii][1] > max_fi:
            f_count = f_count + sort_fri[ii][1]-max_fi
            max_fi = sort_fri[ii][1]
    return f_count
```
