#### 1.1 helper functions

```python
def parselist(inputlist):
    return [tuple([int(x) for x in y.split(",")]) for y in inputlist]
```

```python
def alldistances(junc):
    alldist = {}
    for ii in range(0,len(junc)-1):
        for jj in range(ii+1,len(junc)):
            # print(junc[ii],junc[jj])
            alldist[tuple([junc[ii],junc[jj]])] = sum([(junc[ii][x]-junc[jj][x])**2 for x in range(0,3)])
    return alldist
```

```python
def add_conn(dist,circuits,all_circs):
    if dist[0][0] in all_circs and dist[0][1] in all_circs:
            for circ in circuits:
                if dist[0][0] in circ and dist[0][1] not in circ:
                    for circb in circuits:
                        if dist[0][1] in circb:
                            # print('2',circ,circb)
                            circuits.append(circ.union(circb))
                            circuits.remove(circ)
                            circuits.remove(circb)
                            # print(dist[0],circuits)
                            # print([len(x) for x in circuits],len(all_circs))
                            return circuits,all_circs
    elif dist[0][0] in all_circs or dist[0][1] in all_circs:
        # print(dist[0])
        # print(circuits)
        for circ in circuits:
            if dist[0][0] in circ or dist[0][1] in circ:
                circ.add(dist[0][0])
                circ.add(dist[0][1])
                all_circs.add(dist[0][0])
                all_circs.add(dist[0][1])
    else:
        # print(f"before {circuits}")
            circuits.append(set([dist[0][0],dist[0][1]]))
            all_circs.add(dist[0][0])
            all_circs.add(dist[0][1])
            # print(dist[0])
            # print(f"after {circuits}")

    # print(dist[0],circuits)
    # print([len(x) for x in circuits],len(all_circs))
    return circuits,all_circs
```

#### 1.2 Part 1

```python
def part1(inputlist,n):
    juncs = parselist(inputlist)
    alldist = alldistances(juncs)
    sorted_dist = sorted(alldist.items(), key=lambda kv: (kv[1], kv[0]))[:n]
    circuits =[]    
    all_circs = set()
    # while len(list(all_circs)) < n:
    for dist in sorted_dist:
        # print(dist,all_circs,circuits)
        circuits,all_circs = add_conn(dist,circuits,all_circs)
    conn_prod = 1
    circ_len = [len(x) for x in circuits]
    for circ in sorted(circ_len,reverse=True)[:3]:
        conn_prod = conn_prod*circ
    # print([len(x) for x in circuits])
    return conn_prod
```

```python
def part2(inputlist):
    juncs = parselist(inputlist)
    alldist = alldistances(juncs)
    sorted_dist = sorted(alldist.items(), key=lambda kv: (kv[1], kv[0]))
    circuits =[]    
    all_circs = set()
    for dist in sorted_dist:
        # print(dist,all_circs,circuits)
        circuits,all_circs = add_conn(dist,circuits,all_circs)
        # print(dist[0],len(all_circs))
        if len(all_circs) == len(juncs) and len(circuits) == 1:
            # print(dist[0])
            return dist[0][0][0]*dist[0][1][0]
```
