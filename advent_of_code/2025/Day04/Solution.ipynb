#### 1.1 helper functions

```python
def all_rolls(inputlist):
    roll_list = set([])
    for ii in range(0,len(inputlist)):
        for jj in range(0,len(inputlist[ii])):
            if inputlist[ii][jj]=="@":
                roll_list.add((ii,jj))
    return roll_list
```

```python
def remove_rolls(roll_list):
    new_rolls = set([])
    all_dirs = [
        np.array([-1,-1]),
        np.array([-1,0]),
        np.array([-1,1]),
        np.array([0,-1]),
        np.array([0,1]),
        np.array([1,-1]),
        np.array([1,0]),
        np.array([1,1]),
        ]
    # print(roll_list)
    for roll in roll_list:
        # print(roll)
        n_count = 0
        # print(roll)
        for dir in all_dirs:
            if tuple(np.array(roll)+dir) in roll_list:
                # print(f"Roll {roll}, Dir {dir}")
                n_count = n_count+1
        if n_count >= 4:
            # print(roll)
            new_rolls.add(roll)
    return new_rolls
```

#### 1.2 Part 1

```python
def part1(inputlist):
    roll_set = all_rolls(inputlist)
    all_dirs = [
        np.array([-1,-1]),
        np.array([-1,0]),
        np.array([-1,1]),
        np.array([0,-1]),
        np.array([0,1]),
        np.array([1,-1]),
        np.array([1,0]),
        np.array([1,1]),
        ]
    roll_count = 0
    for roll in roll_set:
        # print(roll)
        n_count = 0
        for dir in all_dirs:
            if tuple(np.array(roll)+dir) in roll_set:
                # if roll in [(0,2),(3,0)]:
                #     print(f"Roll {roll}, Dir {dir}")
                n_count = n_count+1
        if n_count < 4:
            # print(roll)
            roll_count = roll_count+1
    return roll_count
```

#### 1.3 Part 2

```python
def part2(inputlist):
    roll_set = all_rolls(inputlist)
    init_len = len(roll_set)
    new_roll_set = remove_rolls(roll_set)
    while len(roll_set) > len(new_roll_set):
        # print(len(roll_set))
        # print(len(new_roll_set))
        roll_set = new_roll_set
        new_roll_set = remove_rolls(roll_set)
    return init_len-len(roll_set)
```
