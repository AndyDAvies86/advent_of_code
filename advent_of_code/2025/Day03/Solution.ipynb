#### 1.1 helper functions

```python
def parselist(inputlist):
    return [[int(x) for x in y] for y in inputlist]
```

#### 1.2 Part 1

```python
def part1(inputlist):
    batts = parselist(inputlist)
    joltage = 0
    for batt in batts:
        tens = max(batt[:-1])
        tens_pos = batt.index(tens)
        units = max(batt[tens_pos+1:])
        # print(units)
        joltage = joltage + 10*tens+units
    return joltage
```

```python
print(f"Part 1 test is: {part1(test)}")
print(f"Part 1 puzzle is: {part1(start)}")
```

#### 1.3 Part 2

```python
def part2(inputlist):
    batts = parselist(inputlist)
    joltage = 0
    for batt in batts:
        # print(batt)
        batt.append(0)
        for ii in range(0,12):
            # print(ii)
            coeff = max(batt[:-(11-ii+1)])
            coeff_pos = batt.index(coeff)
            joltage = joltage + (10**(11-ii))*coeff
            batt = batt[coeff_pos+1:]
            # print(f"ii{ii}, coeff{coeff}, pos{coeff_pos}, batt{batt}")
    return joltage
```

```python
print(f"Part 2 test is: {part2(test)}")
print(f"Part 2 puzzle is: {part2(start)}")
```
