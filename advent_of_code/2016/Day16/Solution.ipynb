#### 1.1 helper functions

```python
def dragon_step(prior):
    a = prior
    b = prior[::-1]
    b=b.replace('0','2')
    b=b.replace('1','0')
    b=b.replace('2','1')
    return a+'0'+b
```

```python
def checksum(disk):
    disk = str(disk)
    cs_out = ''
    for ii in range(0,len(disk),2):
        if disk[ii] == disk[ii+1]:
            cs_out=cs_out+'1'
        else:
            cs_out=cs_out+'0'
    return cs_out

```

#### 1.2 Part 1

```python
def part1(inputnum,disk_size):
    drag_num = str(inputnum)
    # print(drag_num)
    while len(drag_num) < disk_size:
        drag_num = dragon_step(drag_num)
        # print(drag_num)
    drag_num = drag_num[:disk_size]

    while len(drag_num) % 2 == 0:
        drag_num = checksum(drag_num)

    return drag_num
```

```python
print(f"Part 1 test is: {part1(10000,20)}")
print(f"Part 1 puzzle is: {part1(start,272)}")
```

```python
def part2(inputlist,disk_size):
    return part1(inputlist,disk_size)
```
