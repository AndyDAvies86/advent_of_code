#### 1.1 helper functions

```python
def first4md5(instr):
    MD5hash = hashlib.md5(instr.encode())
    return MD5hash.hexdigest()[:4]
```

```python
def check_doors(instr,pos,pos_dict):
    f4md5=first4md5(instr)
    UDLR = 'UDLR'
    U=1+0*1j
    D=-1+0*1j
    L=0*-1*1j
    R=0+1*1j
    movers = {'U':U,'D':D,'L':L,'R':R}
    for ii in range(0,3):
        if f4md5[ii] in 'bcdef':
            new_pos=movers[UDLR[ii]]+pos
            new_str = instr+UDLR[ii]
            if new_pos.real >= 0 and new_pos.real <= 3 and new_pos.imag >=0 and new_pos.imag <= 3:
                if new_str not in pos_dict:
                    pos_dict[new_pos] = new_str
```
